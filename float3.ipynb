```python
def get_most_pop(users, orders):
    cats = [0] * 13
    
    for id in users.index:
        for cat in range(10, 23):
            if orders[orders['user_id'] == id].iloc[:, cat].max() == 1:
                cats[cat - 10] += 1
                
    return cats
```
